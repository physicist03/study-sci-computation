```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# 데이터 정의
x_data = np.array([0, 176, 295, 392, 500, 610])
y_data = np.array([35, 29, 27, 24, 22, 20])

# 적합시킬 모델 함수 정의
def linear_model(x, m, b):
    return m * x + b

# curve_fit을 이용한 선형회귀 수행
params, covariance = curve_fit(linear_model, x_data, y_data)

# 결과 매개변수 (기울기와 y절편)
m, b = params

# 예측값 계산
y_pred = linear_model(x_data, *params)

# R^2 계산
ss_res = np.sum((y_data - y_pred) ** 2)
ss_tot = np.sum((y_data - np.mean(y_data)) ** 2)
r_squared = 1 - (ss_res / ss_tot)

print(f"기울기: {m}, y절편: {b}")
print(f"R^2: {r_squared}")

# 결과 시각화
plt.scatter(x_data, y_data, label='Data')
plt.plot(x_data, y_pred, label='Fitted line', color='red')
plt.legend()
plt.show()
```
