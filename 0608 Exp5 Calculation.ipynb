```python
import numpy as np
import pandas as pd
import ace_tools as tools

# 주어진 데이터
angles_deg = np.array([15, 30, 45, 60, 75])
N_values = np.array([5, 2, 1, 1, 1])

# 고정 상수
na = 1.000293
t_mm = 0.85
lambda_mm = 0.0006328

# 각도를 라디안으로 변환
angles_rad = np.radians(angles_deg)

# 식에 따라 ng 계산
numerator = na * (1 - np.cos(angles_rad)) * (2 * na * t_mm - N_values * lambda_mm)
denominator = 2 * na * t_mm * (1 - np.cos(angles_rad)) - N_values * lambda_mm
ng_values = numerator / denominator

# 결과를 데이터프레임으로 정리
df = pd.DataFrame({
    "각도 (deg)": angles_deg,
    "무늬수 변화 N": N_values,
    "ng 계산값": ng_values
})

tools.display_dataframe_to_user(name="ng 계산 결과", dataframe=df)
df
```
