```python
a, b, c = 1, 1, -1
while b**2-4*a*c>0:
  xp = (-b+np.sqrt(b**2-4*a*c))/(2*a)
  xm = (-b-np.sqrt(b**2-4*a*c))/(2*a)
  print(f'c={c:.3f}, x={xp:.3f},   {xm:.3f}')
  c += 0.1
```
